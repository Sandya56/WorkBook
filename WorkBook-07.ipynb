### Number is even or odd 
### Number is prime or not

```python
class Numbers():
    def __init__(self,num):
        self.numbers=num
    def is_even(self):
        for i in self.numbers:
            if i%2==0:
                print(f'{i} is even')
            else:
                print(f'{i} is odd')
    def prime_num(self):           
        for i in self.numbers:
            if i>1:
                for j in range(2,i):
                    if i%j==0:
                        print(f'{i} is not prime')
                        break
                    else:print(f'{i} is prime')
                else:continue
n =  Numbers([5,2,6,66,55,101])
n.is_even()
n.prime_num()
```

### Multi level inheritance

```python
class Animal:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def speak(self):
        print(f"{self.name} is speaking")
class Mammal:
    def __init__(self,name,age,colour):
        Animal.__init__(self,name,age)
        self.colour=colour
    def speak(self):
        print(f"{self.name} is making sound")
class Lion:
    def __init__(self,name,age,colour,gender):
        Mammal.__init__(self,name,age,colour)
        self.gender=gender
    def give_details(self):
        print(f"{self.name} is the {self.gender} lion, age of {self.name} is {self.age} and colour {self.colour}")
    def speak(self):
        print(f"{self.name} is roaring")
l=Lion('Simba',56,'Red','Female')
l.give_details()
l.speak()
```
